# Combining and Merging Data Sets

```python
import pandas as pd
import numpy as np
```

## Database Style dataframe merge

```python
df1 = pd.DataFrame({'key': ['bear', 'bear', 'ant', 'cat', 'ant', 'ant', 'bear'], 'data1': range(7)})

df2 = pd.DataFrame({'key': ['ant', 'bear', 'dog'], 'data2': range(3)})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2)
```

```python
pd.merge(df1, df2, on='key')
```

```python
df3 = pd.DataFrame({'lkey': ['bear', 'bear', 'ant', 'cat', 'ant', 'ant', 'bear'], 'data1': range(7)})
df3
```

```python
df4 = pd.DataFrame({'rkey': ['ant', 'bear', 'dog'], 'data2': range(3)})
df4
```

```python
pd.merge(df3, df4, left_on='lkey', right_on='rkey')
```

```python
pd.merge(df1, df2, how='outer')
```

## Merge on Index

```python
left1 = pd.DataFrame({'key': ['ant', 'bear', 'ant', 'ant', 'bear', 'cat'], 'value': range(6)})
```

```python
right1 = pd.DataFrame({'group_val': [3.5, 7]}, index=['ant', 'bear'])
```

```python
pd.merge(left1, right1, left_on='key', right_index=True)
```

```python
pd.merge(left1, right1, left_on='key', right_index=True, how='outer')
```

```python
left = pd.DataFrame({'key1': ['Ohio', 'Ohio', 'Ohio', 'Nevada', 'Nevada'], 'key2': [2000, 2001, 2002, 2001, 2002],
'data': np.arange(5.)})

left
```

```python
right = pd.DataFrame(np.arange(12).reshape((6, 2)), index=[['Nevada', 'Nevada', 'Ohio', 'Ohio', 'Ohio', 'Ohio'],[2001, 2000, 2000, 2000, 2001, 2002]], columns=['event1', 'event2'])

right
```

```python
pd.merge(left, right, left_on=['key1', 'key2'], right_index=True)
```

```python
pd.merge(left, right, left_on=['key1', 'key2'], right_index=True, how='outer')
```
