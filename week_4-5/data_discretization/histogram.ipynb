# Histogram

```python
import matplotlib.pyplot as plt
```

```python
plt.style.use('ggplot')
ages = [12, 13, 15, 19, 20, 20, 24, 26, 32, 39, 49]
plt.hist(ages)
plt.xlabel("Age")
plt.ylabel("Age Disribution")
plt.title("Age Distribution Histogram")
plt.show
```

```python
bins = 5
plt.hist(ages, bins=bins, edgecolor='black')
```

```python
ages = [12, 13, 15, 19, 20, 20, 24, 26, 32, 39, 49]
bins = [10, 15, 20, 25, 30, 35, 40, 45, 50]
plt.hist(ages, bins=bins, edgecolor='black')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as df

sns.set()
h = sns.histplot([1, 5, 9, 4, 7, 11, 14, 17, 13, 18, 19, 31,
                 33, 36, 42, 44, 46, 70, 74, 78, 77], bins=[1, 7, 11, 19, 31, 46, 70, 77])
plt.xlabel('value')
plt.ylabel('frequency')
plt.title('histogram title')

h.set_xticks([13, 35, 50, 67.5])
h.set_xticklabels(["Child", "Young", "Mature", "Old"])

plt.show()
```
