# Matplotlib


Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. 

```python
import pandas as pd
from matplotlib import pyplot as plt
```

## Basics

```python
# Sample data
x = [1, 2, 3]
y = [1, 4, 9]
```

The `title()` function add a title for the plot

the `xlabel()` function adds a label along the x-axis

the `ylabel()` function adds a label along the x-axis

the `legend()` function adds a legend for each line

the `plot()` function accepts variables that will be appear in the plot

```python
plt.plot(x, y)
plt.title("test plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend("y")
plt.show()
```

Add a third variable

```python
z = [10, 5, 0]
```

```python
plt.plot(x, y)
plt.plot(x, z)
plt.title("test plot")
plt.xlabel("x")
plt.ylabel("y and z")
plt.legend(["this is y", "this is z"])
plt.show()
```

## Plot a real world dataset

```python
data = pd.read_csv('countries.csv')
```

```python
data
```

### Compare the population growth in the US and China

```python
us = data[data.country == 'United States']
```

```python
us
```

```python
china = data[data.country == 'China']
```

```python
china
```

### Plot the population growth between US and China

```python
plt.plot(us.year, us.population)
plt.plot(china.year, china.population)
plt.legend(['United States', 'China'])
plt.xlabel('year')
plt.ylabel('population')
plt.show()
```

Divide the population by 1 million

```python
plt.plot(us.year, us.population / 10**6)
plt.plot(china.year, china.population / 10**6)
plt.legend(['United States', 'China'])
plt.xlabel('year')
plt.ylabel('population')
plt.show()
```

```python
us.population
```

Divide the population each year by 100

```python
us.population / us.population.iloc[0] * 100
```

```python
plt.plot(us.year, us.population / us.population.iloc[0] * 100)
plt.plot(china.year, china.population / china.population.iloc[0] * 100)
plt.legend(['United States', 'China'])
plt.xlabel('year')
plt.ylabel('population growth (first year = 100)')
plt.show()
```
